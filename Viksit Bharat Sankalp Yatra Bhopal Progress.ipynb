##### Hamara Sankalp Viksit Bharat is a nationwide campaign to raise awareness through outreach activities to achieve saturation of schemes of Govt. of India across the country covering all Gram Panchayats, Nagar Panchayats and Urban Local Bodies with the objectives of:
* Reaching the unreached - reach out to the vulnerable who are eligible under various schemes but have not availed benefit so far
* Dissemination of information and generating awareness about schemes
* Learning from the citizens - Interaction with beneficiaries of government schemes through their personal stories/ experience sharing
* Enrolment of potential beneficiaries through details ascertained during the Yatra.
The campaign is being taken up by adopting a whole of government approach with active involvement of various Ministries/Departments of Government of India, State Governments, Central Govt. Organizations and Institutions to ensure the widest possible participation of all concerned for the benefit of Citizens of the country and to realize the vision of Viksit Bharat.

##### Here is a Brief Information about Camp Organized in the area of Bhopal Municipal Corporation

```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
```

```python
df = pd.read_excel('VBSY.xlsx')
print('data successfully read into pandas dataframe')
```

```python
df.head()
```

```python
# remove extra space from columns
df.columns = [x.strip() for x in df.columns]
```

```python
df['Camp Location Address'] = df['Camp Location Address'].str.capitalize()
df.head()
```

```python
vidhansabha_wise = df.groupby(['Vidhan Sabha'])['Camp Location Address'].count().to_frame().reset_index()
```

```python
vidhansabha_wise
```

```python
bhopal_progress_df = pd.read_excel('bhopal_progress.xlsx')
print('data successfully read into pandas dataframe')
```

```python
bhopal_progress_df.head()
```

```python
bhopal_progress_df.shape
```

```python
# remove extra space from location column
bhopal_progress_df['Location'] = [x.strip() for x in bhopal_progress_df['Location']]
bhopal_progress_df['Location'] = bhopal_progress_df['Location'].str.capitalize()
```

```python
bhopal_progress_df.head()
```

```python
bhopal_progress_df['Date'] = bhopal_progress_df['Date'].dt.strftime('%d-%m-%Y')
vs_dict = bhopal_progress_df.to_dict(orient='records')
vs_dict[0:3]
```

```python
bhopal_cord = 23.253374, 77.426899
bhopal_map = folium.Map(location=bhopal_cord, zoom_start=12)

camp = plugins.MarkerCluster().add_to(bhopal_map)

for lat, long, row_dict in zip(df['Latitude'], df['Longitude'], vs_dict):
    label = (
        f"VidhanSabha:- {row_dict['Vidhan Sabha']}<br>"
        f"Date:- {row_dict['Date']}<br>"
        f"Location:- {row_dict['Location']}<br>"
        f"Total People Attendance:- {row_dict['Total People Attendance']}"
    )
    
    folium.Marker(
        location=[lat, long],
        icon=None,
        popup=label
    ).add_to(camp)

bhopal_map
```

```python
# save map
bhopal_map.save('VBSY_Bhopal_Camp_Details_Map.html')
```
